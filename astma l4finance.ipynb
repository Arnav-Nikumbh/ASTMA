```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.metrics import classification_report

def analyze_finance_data():
    analyzer = SentimentIntensityAnalyzer()
    
    print("Loading Finance Dataset...")
    try:
        df = pd.read_csv('lab5_sentiment_finance.csv', encoding='latin-1').dropna()
    except FileNotFoundError:
        print("Error: lab5_sentiment_finance.csv not found.")
        return

    def get_vader_label(text):
        score = analyzer.polarity_scores(str(text))['compound']
        if score >= 0.05: return "Positive"
        elif score <= -0.05: return "Negative"
        else: return "Neutral"

    print("Applying VADER Sentiment Analysis...")
    df['vader_prediction'] = df['text'].apply(get_vader_label)
    
    # Capitalize the original category to match VADER's formatting
    df['true_label'] = df['category'].str.capitalize()

    print("\n--- Finance Data Results ---")
    print(df[['text', 'true_label', 'vader_prediction']].head())

    # Generate Classification Report
    print("\n--- Classification Report ---")
    print(classification_report(df['true_label'], df['vader_prediction']))
    

if __name__ == "__main__":
    analyze_finance_data()
```
