```python
import pandas as pd
import networkx as nx

def run_lab7_higgs():
    print("Loading Higgs Twitter Dataset...")
    df = pd.read_csv('lab8_higgs-social_network.csv')
    
    # Create directed graph
    print("Building Directed Graph...")
    G = nx.from_pandas_edgelist(df, 'source', 'target', create_using=nx.DiGraph())
    print(f"Nodes: {G.number_of_nodes()} | Edges: {G.number_of_edges()}")
    
    # Apply PageRank
    print("Computing PageRank (alpha=0.85)...")
    pagerank = nx.pagerank(G, alpha=0.85)
    
    # Display ranking
    sorted_pr = sorted(pagerank.items(), key=lambda x: x[1], reverse=True)
    print("\n--- Top 10 Higgs Nodes by PageRank ---")
    for page, rank in sorted_pr[:10]:
        print(f"Node {page} : {round(rank, 6)}")

if __name__ == "__main__":
    run_lab7_higgs()
```
