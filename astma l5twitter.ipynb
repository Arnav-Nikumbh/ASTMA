```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.metrics import classification_report

def analyze_twitter_data():
    analyzer = SentimentIntensityAnalyzer()
    
    print("Loading Twitter Dataset...")
    try:
        df = pd.read_csv('lab5_Twitter_Data_sentiment.csv', encoding='latin-1').dropna()
    except FileNotFoundError:
        print("Error: lab5_Twitter_Data_sentiment.csv not found.")
        return

    # Map Original Categories to Text
    category_map = {-1.0: 'Negative', 0.0: 'Neutral', 1.0: 'Positive'}
    df['true_label'] = df['category'].map(category_map)

    def get_vader_label(text):
        score = analyzer.polarity_scores(str(text))['compound']
        if score >= 0.05: return "Positive"
        elif score <= -0.05: return "Negative"
        else: return "Neutral"

    print("Applying VADER Sentiment Analysis...")
    df['vader_prediction'] = df['text'].apply(get_vader_label)

    print("\n--- Twitter Data Results ---")
    print(df[['text', 'true_label', 'vader_prediction']].head())
    
    # Generate Classification Report
    print("\n--- Classification Report ---")
    # Drop any rows where true_label might be NaN due to mapping issues
    valid_df = df.dropna(subset=['true_label', 'vader_prediction'])
    print(classification_report(valid_df['true_label'], valid_df['vader_prediction']))
    
    
if __name__ == "__main__":
    analyze_twitter_data()
```
