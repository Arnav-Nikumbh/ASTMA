```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import pairwise_distances
import pandas as pd
```

```python
d0 = 'This is a sample notebook'
d1 = 'This notebook is sample'
d2 = 'This is another sample notebook'
string = [d0, d1, d2]
```

TF - IDF

```python
tfidf = TfidfVectorizer()
result = tfidf.fit_transform(string)
print('\nidf values:')
for ele1, ele2 in zip(tfidf.get_feature_names_out(), tfidf.idf_):
    print(ele1, ':', ele2)
```

```python
print('\nWord indexes:')
print(tfidf.vocabulary_)
print('\ntf-idf value:')
print(result)
print('\ntf-idf values in matrix form:')
print(result.toarray())
```

```python
count_vectorizer = CountVectorizer()
sparse_matrix = count_vectorizer.fit_transform(string)
```

```python
doc_term_matrix = sparse_matrix.todense()
df = pd.DataFrame(doc_term_matrix, 
                  columns=count_vectorizer.get_feature_names_out(), 
                  index=['d0', 'd1', 'd2'])

print("Document-Term Matrix:\n", df)
print("-" * 30)
```

```python
cosine_sim = cosine_similarity(sparse_matrix)
sim_df = pd.DataFrame(cosine_sim, 
                      columns=['d0', 'd1', 'd2'], 
                      index=['d0', 'd1', 'd2'])

print("Cosine Similarity Matrix:\n", sim_df)
```

```python
vectorizer = CountVectorizer(binary=True, token_pattern=r"(?u)\b\w+\b")
X = vectorizer.fit_transform(string)
```

```python
jaccard_distance = pairwise_distances(X.toarray(), metric='jaccard')
```

```python
jaccard_similarity = 1 - jaccard_distance
```

```python
df = pd.DataFrame(jaccard_similarity, 
                  index=['d0', 'd1', 'd2'], 
                  columns=['d0', 'd1', 'd2'])

print(df)
```
