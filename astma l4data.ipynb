```python

import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.metrics import classification_report

def analyze_general_sentiment_data():
    analyzer = SentimentIntensityAnalyzer()
    
    print("Loading General Sentiment Dataset...")
    try:
        df = pd.read_csv('lab5_sentiment_data.csv', encoding='latin-1').dropna()
    except FileNotFoundError:
        print("Error: lab5_sentiment_data.csv not found.")
        return

    # Map the numeric sentiment (Assuming 0: Negative, 1: Neutral, 2: Positive)
    sentiment_map = {0: 'Negative', 1: 'Neutral', 2: 'Positive'}
    df['true_label'] = df['Sentiment'].map(sentiment_map)

    def get_vader_label(text):
        score = analyzer.polarity_scores(str(text))['compound']
        if score >= 0.05: return "Positive"
        elif score <= -0.05: return "Negative"
        else: return "Neutral"

    print("Applying VADER Sentiment Analysis...")
    df['vader_prediction'] = df['text'].apply(get_vader_label)

    print("\n--- General Sentiment Data Results ---")
    print(df[['text', 'true_label', 'vader_prediction']].head(20))
    
    # Generate Classification Report
    print("\n--- Classification Report ---")
    valid_df = df.dropna(subset=['true_label', 'vader_prediction'])
    print(classification_report(valid_df['true_label'], valid_df['vader_prediction']))
    
    

if __name__ == "__main__":
    analyze_general_sentiment_data()
```
